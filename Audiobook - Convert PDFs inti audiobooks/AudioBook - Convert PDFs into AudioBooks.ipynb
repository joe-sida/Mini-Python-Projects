```python
# Import all neccessary modules

from gtts import gTTS
import PyPDF3
```

```python
# Open the pdf file
book = open('Franz Kafka_The Trial.pdf', 'rb')

#Create and object for the pdf file
reader = PyPDF3.PdfFileReader(book)

#count the number of pages
page_count = reader.numPages

#Extract the text from the pdf 
text_list = []

for i in range (page_count):
    try:
        page = reader.getPage(i)
        text_list.append(page.extractText())
    except:
        pass
        
```

```python
#Join the text into one line string
full_text_string = " ".join(text_list)
print(full_text_string)
```

```python
#Call the Google text to speech library(gTTS) and create the audio book
language = "en-us"
audio_file = gTTS(text=full_text_string, lang=language, slow=False)
audio_file.save("AudioBook_2.mp3")
```
